
1. **PG-13 (Parents Strongly Cautioned)**: Some material may be inappropriate for children under 13. Parents are urged to be cautious about the content and consider whether it is suitable for their children.

2. **TV-MA (Mature Audiences)**: This program is specifically designed to be viewed by adults and is not suitable for children under 17. It may contain explicit content, such as strong language, sexual content, or graphic violence.

3. **TV-PG (Parental Guidance Suggested)**: Some material may not be suitable for children. Parents are urged to provide "parental guidance," and the content may contain infrequent mild language, some violence, or suggestive dialogue.

4. **TV-14 (Parents Strongly Cautioned - Ages 14 and above)**: This program may be unsuitable for children under 14 years of age. Parents are strongly cautioned, as it may contain intense violence, sexual content, or strong language.

5. **TV-Y7 (Directed to Older Children - Ages 7 and above)**: This program is designed for children aged 7 and above. It may be more intense than content designated as TV-Y and may contain fantasy violence or comedic violence.

6. **TV-Y (All Children)**: This program is suitable for all children. It contains little or no violence, no strong language, and little or no sexual content.

7. **TV-G (General Audience)**: All ages are admitted. It is specifically designed for children and may be suitable for all audiences.

8. **PG (Parental Guidance Suggested)**: Some material may not be suitable for children. Parents are urged to provide "parental guidance." This may include some mild language, thematic elements, or brief violence.

9. **R (Restricted)**: Restricted to viewers over the age of 17 or 18. Parents are strongly urged to learn more about the movie before taking their children to see it. It may contain adult themes, adult activity, hard language, intense or persistent violence, sexually-oriented nudity, drug abuse, or other elements.

10. **G (General Audience)**: All ages are admitted. It is specifically designed for children and may be suitable for all audiences.

11. **NC-17 (Adults Only)**: No one 17 and under admitted. It is similar to an R rating but indicates a stricter stance on age restrictions.

12. **NR (Not Rated)**: This film has not been submitted for a rating or is an uncut version of a film that was submitted.

13. **TV-Y7-FV (Directed to Older Children - Fantasy Violence)**: This program is designed for children aged 7 and above and may contain fantasy violence.

14. **UR (Unrated)**: This film has not been submitted for a rating or is an uncut version of a film that was submitted.

These ratings are provided by organizations like the Motion Picture Association (MPA) for movies and the TV Parental Guidelines for television shows. They aim to assist viewers in making informed decisions about the content they consume based on their own preferences and the age appropriateness for their audience.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')  
plt.show()
```

```python
release_year_counts = df['release_year'].value_counts()
print(release_year_counts)
```

```python
plt.figure(figsize=(15, 6))


sns.countplot(x='release_year', data=df, palette='viridis',hue='release_year', legend=False) 

# Customize the plot
plt.title('Distribution of Movies across Years')
plt.xlabel('Year')
plt.ylabel('Count')


plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_categories = df['listed_in'].value_counts()
print(top_categories)
```

```python
plt.figure(figsize=(15, 6))


sns.countplot(x='country', data=df, palette='viridis',hue='country', legend=False) 

# Customize the plot
plt.title('Distribution of Movies of different countries')
plt.xlabel('Country')
plt.ylabel('Count')


plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Monthly addition of content
df = df.copy()

# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract the month and create a new column
df['month_added'] = df['date_added'].dt.month

# Count the occurrences of each month
monthly_addition = df['month_added'].value_counts().sort_index()

# Print the result
print(monthly_addition)
```
